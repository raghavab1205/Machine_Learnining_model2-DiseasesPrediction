```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

#Reading dataset

```python
# Separate features and target
data_train = pd.read_csv("Training_diseases.csv")
data_train.columns
```

```python
data_train.head()
```

```python
data_train.describe()
```

#Cleaning dataset

```python
data_train.isnull().sum()
```

#input and output training dataset

```python
X_train = data_train.drop(['prognosis','Unnamed: 133'], axis=1)
y_train = data_train['prognosis']
```

#input and output test dataset

```python
data_test = pd.read_csv("Testing_dieseases.csv")
data_test.head()
```

```python
data_test.columns
```

```python
X_test = data_test.drop(['prognosis'], axis=1)
y_test = data_test['prognosis']
```

#Training the model

```python
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train,y_train)
```

#Predicting the diseases

```python
y_pred = rf_classifier.predict(X_test)
y_pred
```

#test result-Diseases

```python
y_test
```

#checking accuracy and getting Classification report

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
